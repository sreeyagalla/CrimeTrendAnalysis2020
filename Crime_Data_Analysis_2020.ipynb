```python
#Performing test to determine p-value of the data
from statsmodels.tsa.stattools import adfuller
def adf_test(series):
    result=adfuller(series)
    print('ADF Statistics: {}'.format(result[0]))
    print('p- value: {}'.format(result[1]))
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis, reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("weak evidence against null hypothesis, time series has a unit root")
adf_test(df1['No Of Crimes'])
```

```python
#Choosing the optimal parameters for ARIMA model
from pmdarima import auto_arima
import statsmodels.api as sm
model=auto_arima(df1['No Of Crimes'], trace=True)
model.summary()
```

```python
#Preparing training and test dataset
print(df1.shape)
train=df1.iloc[:-30]
test=df1.iloc[-30:]
print(train.shape,test.shape)
print(df1.shape)
train=df1.iloc[:-30]
test=df1.iloc[-30:]
print(train.shape,test.shape)
print(test.iloc[0],test.iloc[-1])
```

```python
#Model Training
model = sm.tsa.arima.ARIMA(df1['No Of Crimes'], order=(5,1,3))
model=model.fit()
model.summary()
```

```python
#Checking Test dataset
start=len(train)
end=len(train)+len(test)-1
pred=model.predict(start=start,end=end,typ='levels').rename('ARIMA Predictions')
pred.plot(legend=True)
test['No Of Crimes'].plot(legend=True)
plt.title("Test model accuracy")
plt.ylabel("Crime Rate")
```

```python
#RMSE value
from sklearn.metrics import mean_squared_error
from math import sqrt
test['No Of Crimes'].mean()
rmse=sqrt(mean_squared_error(pred,test['No Of Crimes']))
print(rmse)
```

**Correlation with Economic Factors**

```python
#5
economic_data = pd.read_csv('Data.csv')
economic_data['DATE'] = pd.to_datetime(economic_data['DATE'])
economic_data['YEAR'] = economic_data['DATE'].dt.year
economic_data['MONTH'] = economic_data['DATE'].dt.month
economic_data.dropna(subset=['MED HOUSEHOLD INCOME'], inplace=True)
economic_data = economic_data.drop(columns=['DATE'])
crime_counts_by_month = df.groupby(['Year', 'Month']).size().reset_index(name='Crime_Counts')
crime_counts_by_month = crime_counts_by_month.reset_index(drop=True)
merged_data = pd.merge(crime_counts_by_month, economic_data, left_on=['Month','Year'],
                       right_on=['MONTH','YEAR'], how='inner')
correlation_results = {}
for factor in ['UNRATE(%)', 'CONSUMER CONF INDEX', 'PPI-CONST MAT.', 'CPIALLITEMS', 'INFLATION(%)',
               'MORTGAGE INT. MONTHLY AVG(%)', 'MED HOUSEHOLD INCOME', 'CORP. BOND YIELD(%)',
               'GDP PER CAPITA']:
    correlation = crime_counts_by_month['Crime_Counts'].corr(merged_data.groupby('MONTH')[factor].sum(), method='pearson')
    correlation_results[factor] = correlation

print("Correlations:")
for factor, correlation in correlation_results.items():
    print(f"{factor}: {correlation}")
```

```python
factors = list(correlation_results.keys())
correlations = list(correlation_results.values())


plt.figure(figsize=(10, 6))
plt.barh(factors, correlations, color='skyblue')
plt.title('Pearson Correlations Between Economic Factors and Monthly Crime Counts')
plt.xlabel('Pearson Correlation Coefficient')
plt.ylabel('Economic Factors')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```

 **Predicting Future Trends**

```python
from statsmodels.tsa.arima.model import ARIMA
from datetime import timedelta

df['DATE_OCC'] = pd.to_datetime(df['DATE_OCC'])
df.set_index('DATE_OCC', inplace=True)
daily_crime_counts = df['DR_NO'].resample('D').count()

train_data = daily_crime_counts['2020-01-01':'2023-09-30']
test_data = daily_crime_counts['2023-10-01':]

model = ARIMA(train_data, order=(5, 1, 0))
model_fit = model.fit()

start_date = test_data.index[0]
end_date = test_data.index[-1] + timedelta(days=1)
forecast = model_fit.predict(start=start_date, end=end_date, typ='levels')

plt.figure(figsize=(12, 6))
plt.plot(train_data, label='Training Data')
plt.plot(test_data, label='Test Data')
plt.plot(forecast, label='ARIMA Forecast', color='red')
plt.title('ARIMA Forecast for Crime Trends')
plt.xlabel('Date')
plt.ylabel('Number of Crimes')
plt.legend()
plt.show()
```
